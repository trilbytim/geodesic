```python
import pandas
df = pandas.read_csv(open("geodesicspraylines.csv", "r"))
dfI = pandas.read_csv(open("geodesicspraylinesI.csv", "r"))
df2 = pandas.read_csv(open("geodesicspraylines2.csv", "r"))
dfI
```

```python
#df.loc[:,["pathlength"]]
df.columns
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
pathlength = list(df.pathlength)
cylangleadvance = list(df.cylangleadvance)
for i in range(len(pathlength)):
    if cylangleadvance[i] == -1:
        pathlength[i] = 0 
plt.plot(df.angleout, pathlength)
plt.xlabel("angle out")
plt.ylabel("path length")
```

```python
cylangleadvance = list(df.cylangleadvance)
for i in range(len(cylangleadvance)):
    if cylangleadvance[i] != -1 and cylangleadvance[i] < 0.0:
        cylangleadvance[i] += 360
plt.figure(figsize=(15,5))
#plt.plot(df.angleout, cylangleadvance)
plt.plot(df.angleout, -df.windingrot, label="winding")
plt.plot(dfI.angleout, -dfI.windingrot, label="winding tang")
plt.plot(df2.angleout, -df2.windingrot)
plt.xlabel("angle out")
plt.ylabel("cycle advance")
plt.legend()
```
