```python
from matplotlib import pyplot as plt
%matplotlib inline
pathlength = list(df.pathlength)
for i in range(len(pathlength)):
    if cylangleadvance[i] == -1:
        pathlength[i] = 0 
plt.plot(df.angleout, pathlength)
plt.xlabel("angle out")
plt.ylabel("path length")
```

```python
cylangleadvance = list(df.cylangleadvance)
for i in range(len(cylangleadvance)):
    if cylangleadvance[i] != -1 and cylangleadvance[i] < 0.0:
        cylangleadvance[i] += 360
plt.plot(df.angleout, cylangleadvance)
plt.xlabel("angle out")
plt.ylabel("cycle advance")
```
